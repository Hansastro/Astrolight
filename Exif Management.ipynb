Exif Tags:
- Exposure Time: 0x829a
- F Stop: 0x829d
- ISO: 0x8827
- Shutter Speed: 0x9201
- Aperture value: 0x9202
- Focal Length: 0x920a

```python
ExifField = (('Exposure time', 0x829a),
            ('F Stop', 0x829d),
            ('ISO', 0x8827),
            ('Shutter Speed', 0x9201),
            ('Aperture value', 0x9202),
            ('Focal Length', 0x920a))
for f in ExifField:
    print('{}: {}'.format(f[0], exifList['Exif'][f[1]]))
```
